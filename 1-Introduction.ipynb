#### SciPy
SciPy is a collection of functions for scientific computing in Python

```python
from scipy import sparse

data=np.eye(5)

print("Numpy array:\n{}".format(data))
```

```python
sparse_matrix=sparse.csc_matrix(data)
print("\nSciPy sparse CSR matrix:\n{}".format(sparse_matrix))
```

```python
data=np.ones(10)
row_indices=np.arange(10)
col_indices=np.arange(10)
eye_coo=sparse.coo_matrix((data,(row_indices,col_indices)))
print("COO representation:\n{}".format(eye_coo))
```

#### Measuring Success: Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
#X_train,X_test,y_train,y_test=train_test_split(iris_dataset['data'],iris_dataset['target'],test_size=0.3,random_state=0)
```

```python
print("X_train shape:",X_train.shape) 
print("y_train shape:",y_train.shape)
```

```python
print('X_test shape:',X_test.shape)
print('y_test shape:',y_test.shape)
```

#### First Things First: Look at Your Data

```python
import mglearn
# create dataframe from data in X_train=X_train
# label the columns using the strings in iris_dataset.feature_names = columns=iris_dataset.feature_names
iris_dataframe=pd.DataFrame(X_train,columns=iris_dataset.feature_names)
# create a scatter matrix from the dataframe, color by y_train
graph=pd.plotting.scatter_matrix(iris_dataframe,c=y_train,figsize=(10,10),marker='*',hist_kwds={'bins':15},s=50,alpha=0.8,cmap=mglearn.cm3)
```

#### Building Your First Model: k-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#object create
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

#### Making Predictions

```python
X_new=np.array([[5,2.9,1,2.0]])
print('X_new.shape: {}'.format(X_new.shape))
```

```python
predict=knn.predict(X_new)
print("Prediction:",predict)
```

```python
print('predicted target name:',iris_dataset['target_names'][predict])
```

#### Evaluating the Model

```python
y_pred=knn.predict(X_test)
print('Test set prediction:\n{}'.format(y_pred))
```

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n", y_pred)
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
